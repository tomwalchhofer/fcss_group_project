```python
import pandas as pd
```

```python
def combine_party_df(party: str,
             df1: pd.DataFrame,
             df2: pd.DataFrame) -> pd.DataFrame:

    df = pd.concat([df1, df2], axis=0)
    df['Datum'] = pd.to_datetime(df['Datum'], errors='coerce', utc=True)
    df['Datum'] = df['Datum'].dt.tz_localize(None)
    df = df.sort_values('Datum', inplace=False)
    df['Partei'] = party
    df.set_index('Datum', inplace=True)

    return df
```

```python
df_fpoe_parlament = pd.read_csv('parlament/ots_scraper_data_fpoe_parlament.csv')
display(df_fpoe_parlament)
```

```python
df_fpoe_partei = pd.read_csv('partei/ots_scraper_data_fpoe_partei.csv')
display(df_fpoe_partei)
```

```python
fpoe = combine_party_df(party='FPÖ',
                           df1=df_fpoe_parlament,
                           df2=df_fpoe_partei)

fpoe = fpoe.dropna() # 0 rows total
display(fpoe)
print(fpoe.shape)
print(fpoe.index.dtype)
```

```python
fpoe.to_csv('fpoe.csv')
```
