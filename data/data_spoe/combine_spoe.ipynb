```python
import pandas as pd
```

```python
def combine_party_df(party: str,
             df1: pd.DataFrame,
             df2: pd.DataFrame) -> pd.DataFrame:

    df = pd.concat([df1, df2], axis=0)
    df['Datum'] = pd.to_datetime(df['Datum'], errors='coerce', utc=True)
    df['Datum'] = df['Datum'].dt.tz_localize(None)
    df = df.sort_values('Datum', inplace=False)
    df['Partei'] = party
    df.set_index('Datum', inplace=True)

    return df
```

```python
df_spoe_parlament = pd.read_csv('parlament/ots_scraper_data_spoe_parlament.csv')
display(df_spoe_parlament)
```

```python
df_spoe_partei = pd.read_csv('partei/ots_scraper_data_spoe_partei.csv')
display(df_spoe_partei)
```

```python
spoe = combine_party_df(party='SPÖ',
                           df1=df_spoe_parlament,
                           df2=df_spoe_partei)

spoe = spoe.dropna() # 18 rows total
display(spoe.tail(40))
print(spoe.shape)
print(spoe.index.dtype)
```

```python
spoe.to_csv('spoe.csv')
```
