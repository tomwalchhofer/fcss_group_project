```python
import pandas as pd
```

```python
def combine_party_df(party: str,
             df1: pd.DataFrame,
             df2: pd.DataFrame) -> pd.DataFrame:

    df = pd.concat([df1, df2], axis=0)
    df['Datum'] = pd.to_datetime(df['Datum'], errors='coerce', utc=True)
    df['Datum'] = df['Datum'].dt.tz_localize(None)
    df = df.sort_values('Datum', inplace=False)
    df['Partei'] = party
    df.set_index('Datum', inplace=True)

    return df
```

```python
df_gruene_parlament = pd.read_csv('parlament/ots_scraper_data_gruene_parlament.csv')
display(df_gruene_parlament)
```

```python
df_gruene_partei = pd.read_csv('partei/ots_scraper_data_gruene_partei.csv')
display(df_gruene_partei)
```

```python
gruene = combine_party_df(party='Grüne',
                           df1=df_gruene_parlament,
                           df2=df_gruene_partei)

gruene = gruene.dropna() # 1 row total
display(gruene)
print(gruene.shape)
print(gruene.index.dtype)
```

```python
gruene.to_csv('gruene.csv')
```
