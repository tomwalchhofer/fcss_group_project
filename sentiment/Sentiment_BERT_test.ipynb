```python
import pandas as pd
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="oliverguhr/german-sentiment-bert")
```

```python
# Beispieltext um Funktion des Tools zu testen
texts = ["Das ist ein toller Tag!", "Ich bin enttäuscht von dem Ergebnis."]

print(texts[0])
sentiment_dict = pipe(texts[0])
print(sentiment_dict)
    
print(texts[1])
sentiment_dict = pipe(texts[1])
print(sentiment_dict)
```

```python
df_oevp = pd.read_csv("press_releases_oevp.csv")
display(df_oevp)
```

```python
df_oevp['BERT_class'] = "Neutral"
for i in range(len(df_oevp["Content"])):
    sentiment_dict = pipe(df_oevp["Content"].iloc[i])[0]
    df_oevp.loc[i, 'BERT_class'] = sentiment_dict["label"].capitalize()
```
