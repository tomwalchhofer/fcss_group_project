```python
import pandas as pd
from germansentiment import SentimentModel
```

```python
model = SentimentModel()
```

```python
def sentiment_calculator(text):
    classes, probabilities = model.predict_sentiment([text], output_probabilities = True)
    weighted_sum = (probabilities[0][0][1])*3+(probabilities[0][1][1])*1+(probabilities[0][2][1])*2
    max_weighted_sum = 3 + 2 + 1
    min_weighted_sum = 0
    normalised_score = 2 * (weighted_sum - min_weighted_sum) / (max_weighted_sum - min_weighted_sum) - 1
    return normalised_score
```

```python
def sort_angelobung(datum: str, data: pd.DataFrame):
    datum = pd.to_datetime(datum)
    df_bevor = data[data.index < datum]
    df_nacher = data[data.index > datum]

    return df_bevor, df_nacher
```

```python
fpoe_gesamt = pd.read_csv('data/data_fpoe/fpoe.csv', parse_dates=[0], index_col=0)
gruene_gesamt = pd.read_csv('data/data_gruene/gruene.csv', parse_dates=[0], index_col=0)
spoe_gesamt = pd.read_csv('data/data_spoe/spoe.csv', parse_dates=[0], index_col=0)
```

```python
fpoe_gesamt["Sentiment Score normalised"]=fpoe_gesamt["Inhalt"].apply(sentiment_calculator)
display(fpoe_gesamt)
```

```python
gruene_gesamt["Sentiment Score normalised"]=gruene_gesamt["Inhalt"].apply(sentiment_calculator)
display(gruene_gesamt)
```

```python
spoe_gesamt["Sentiment Score normalised"]=spoe_gesamt["Inhalt"].apply(sentiment_calculator)
display(spoe_gesamt)
```

```python
fpoe_davor_n, fpoe_danach_n = sort_angelobung(datum='2017-12-18', data=fpoe_gesamt)
display(fpoe_davor_n)
display(fpoe_danach_n)
```

```python
gruene_davor_n, gruene_danach_n = sort_angelobung(datum='2020-01-07', data=gruene_gesamt)
display(gruene_davor_n)
display(gruene_danach_n)
```

```python
spoe_davor_n, spoe_danach_n = sort_angelobung(datum='2007-01-11', data=spoe_gesamt)
display(spoe_davor_n)
display(spoe_danach_n)
```
